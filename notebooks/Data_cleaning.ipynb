#### Now we will start the Data cleaning process

- first we will drop the unnecesssary column

```python
df1.drop(columns=['Unnamed: 0'],inplace=True)
```

- Now we chaeck for the duplicate product name

```python
duplicate=df1['name'].duplicated().sum()
# now we will print the duplicate names
print(f"Duplicate Product Names: {duplicate}")
```

- Now we will inspect the duplicate values

```python
# jsut displaying the names 
duplicate_products = df[df["name"].duplicated(keep=False)]

duplicate_products.head(10)
```

##### Observation & Decision
- The dataset contains 155,375 duplicate product names, but no duplicate rows.

- Therefore, duplicate product names were retained for building the recommendation system.

- Now we will analysis the missing values

```python
# Count of thr missing values in our data
df1.isnull().sum()
```

```python
# Now we will create a missing a value summary table
missing_df = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round((df.isnull().sum()/len(df))*100,2)})

missing_df = missing_df[missing_df["Missing Values"] > 0]
missing_df.sort_values(by="Percentage", ascending=False)
```

- Now we will clean the missing values

```python
# just checking the unique values in the rating column
df1["ratings"].unique()[:20]
```

```python
# now we are Checking  if ratings and no_of_ratings are missing in the same rows
(df1["ratings"].isna() == df["no_of_ratings"].isna()).all()
```

```python
# Convert ratings column from object to float
df1["ratings"] = pd.to_numeric(df1["ratings"], errors="coerce")
```

```python
# now we calculate the median in the rating column
median_rating = df1["ratings"].median()
print("Median Rating:", median_rating)
```

```python
# now we will fill the missing values using the median
df1["ratings"] = df1["ratings"].fillna(median_rating)
```

```python
# now all the missing values are filled
df1["ratings"].isnull().sum()
```

##### Observation & Decision
- The ratings column contained 31.87% missing values.
- The column was converted to a numeric data type before imputation.
- Missing values were filled using the median because it is robust to skewness and outliers.
- This preserved all products while preparing the feature.

#### Now we will work on the no of rating

```python
print(df1["no_of_ratings"].dtype)
```

```python
# science the rating columns has commas we will remove them now
df1["no_of_ratings"] = df1["no_of_ratings"].str.replace(",", "", regex=False)
```

```python
# now we are converting the datatype from object to float
df1["no_of_ratings"] = pd.to_numeric(df1["no_of_ratings"],errors="coerce")
```

```python
print(df1["no_of_ratings"].dtype)
```

```python
# now we will chrck the missing values
df1["no_of_ratings"].isnull().sum()
```

```python
# distribution of the no of ratings

plt.figure(figsize=(10,5))
sns.histplot(df1["no_of_ratings"], bins=50, kde=True,color='slategray')
plt.title("Distribution of Number of Ratings")
plt.show()
```

```python
# now  calculate median 
median_reviews = df1["no_of_ratings"].median()
print("Median:", median_reviews)
```

```python
# Fill missing values with median
df1["no_of_ratings"] = df1["no_of_ratings"].fillna(median_reviews)
```

##### Observation & Decision
- The no_of_ratings column showed a highly right-skewed distribution with a few products having extremely large review counts.
- The mean (840.78) was much higher than the median (20), indicating the presence of significant outliers.
- Therefore, missing values were imputed using the median, as it better represents the typical product and is less affected by extreme values.
- This approach preserves the distribution while preparing the feature for the hybrid recommendation system.

##### Now we will clean the Discount price column

```python
# checking the datatype
print(df1["discount_price"].dtype)
```

```python
# before converting the datatype we have to remove ₹ and comma from the column
df1["discount_price"] = (df1["discount_price"].str.replace("₹", "", regex=False).str.replace(",", "", regex=False))
```

```python
# now we will convert the datatype to the float
df1["discount_price"] = pd.to_numeric(df1["discount_price"],errors="coerce")
```

```python
df1["discount_price"].describe()
```

```python
# now we will see the distrubution of the data 
plt.figure(figsize=(10,5))

sns.histplot(df1["discount_price"], bins=50, kde=True,color='slategray')
plt.title("Distribution of Discount Price")
plt.xlabel("Discount Price")
plt.ylabel("Count")
plt.show()
```

```python
# again calculate median
median_discount_price = df1["discount_price"].median()
print("Median Discount Price:", median_discount_price)
```

```python
# Fill missing values
df1["discount_price"] = df1["discount_price"].fillna(median_discount_price)
```

##### Observation & Decision
- The discount_price column showed a highly right-skewed distribution with a few extremely expensive products.
- The mean (₹2,623.16) was much higher than the median (₹679), indicating the presence of significant outliers.
- Therefore, missing values were imputed using the median, as it better represents the typical product price and is less affected by extreme values.
- This approach preserves the price distribution.

#### Now we will work on the actual price column 

```python
# same we first remove the ₹ and the comma
df1["actual_price"] = (df1["actual_price"].str.replace("₹", "", regex=False).str.replace(",", "", regex=False))
```

```python
# now we will convert them to float datatype
df1["actual_price"] = pd.to_numeric(df1["actual_price"],errors="coerce")
```

```python
#just analysis the distrubution
df1["actual_price"].describe()
```

```python
# just ploting the distrubution of data 
plt.figure(figsize=(10,5))

sns.histplot(df1["actual_price"], bins=50, kde=True,color='slategray')
plt.title("Distribution of Actual Price")
plt.xlabel("Actual Price")
plt.ylabel("Count")
plt.show()
```

```python
# the max price is to large so we will check are them
df1.nlargest(10, "actual_price")[["name", "actual_price", "discount_price"]]
```

```python
# now we will see how many product have too large price
df1[df1["actual_price"] > 1000000].shape
```

```python
# Replace impossible prices (greater than ₹50 lakh) with NaN
df1.loc[df1["actual_price"] > 5000000, "actual_price"] = np.nan
```

- Why we are replacing with ₹50 lakh :-
- The Samsung TV (₹16 lakh) is expensive but we can  believe on that .
- same with the Luxury diamonds around ₹10–12 lakh are also can belie on that .
- but ₹6.1 crore and ₹990 crore are almost certainly data errors.

```python
# we are just verifying 
df1.nlargest(10, "actual_price")[["name", "actual_price", "discount_price"]]
```

```python
# we will do the same median imputation
median_actual_price = df1["actual_price"].median()
df1["actual_price"] = df1["actual_price"].fillna(median_actual_price)
```

```python
# jsut verifying is there any missing value now 
print(df1["actual_price"].isnull().sum())
```

##### Observation & Decision

- The actual_price column contained two unrealistic values (₹61,082,990 and ₹9,900,000,000)
- These values were treated as invalid and replaced with NaN before imputation.
- Missing and invalid values were then filled using the median, as the price distribution was also highly right-skewed.
