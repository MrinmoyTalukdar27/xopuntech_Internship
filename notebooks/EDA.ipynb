##### now we will see distribution of Product Ratings

```python
#ploting the data distrubution
plt.figure(figsize=(10, 8))

sns.countplot(x="ratings",data=df1,order=sorted(df1["ratings"].unique()),color="slategray")
plt.title("Distribution of Product Ratings", fontsize=14)
plt.xlabel("Ratings")
plt.ylabel("Number of Products")
plt.xticks(rotation=90)
plt.show()
```

##### Observation (for your notebook)
- Product ratings range from 1.0 to 5.0, with most ratings concentrated between 3.5 and 4.5.
- A very large peak is observed at 3.9, which is expected because missing values were imputed using the median rating during data cleaning.
- Very few products have ratings below 3.0, indicating that low-rated products are relatively uncommon.
- Overall, the dataset is dominated by positively rated products.

##### Now we do the correlation analysis

```python
# jsut ploting a heatmap too see the corelastion between numarical columns 
plt.figure(figsize=(8,4))

correlation = df1[["ratings","no_of_ratings","discount_price","actual_price"]].corr()
sns.heatmap(correlation,annot=True,cmap="Blues",fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

##### Observation
- discount_price and actual_price show a strong positive correlation (0.93), indicating that products with higher original prices also tend to have higher discounted prices.
- ratings have a very weak correlation with both prices and the number of ratings, suggesting that customer ratings are largely independent of product price.
- no_of_ratings also has almost no correlation with product prices, meaning popular products are not necessarily more expensive.
- Overall, the dataset shows low multicollinearity, except for the strong relationship between discount_price and actual_price.

##### Now we will check the relationship between the actual price and the discuounted price

```python
# just scatterplot to find the relation
plt.figure(figsize=(8,4))

sns.scatterplot(data=df1,x="actual_price",y="discount_price", color='slategray',alpha=0.4)
plt.title("Actual Price vs Discount Price")
plt.xlabel("Actual Price (₹)")
plt.ylabel("Discount Price (₹)")
plt.show()
```

##### Observation
- A strong positive relationship exists between actual_price and discount_price.
- As the actual price increases, the discounted price also tends to increase.
- Most products are concentrated in the lower price range, while only a few premium products appear at higher price levels.
- A small number of high-priced products act as outliers, but they follow the same overall positive trend.

##### Now we will find the Ratings vs Number of Ratings

```python
# we are ploting a scetterplot to see  dose products with higher ratings also receive more customer ratings
plt.figure(figsize=(8,4))

sns.scatterplot(data=df1,x="ratings",y="no_of_ratings",alpha=0.4,color='slategray')
plt.title("Ratings vs Number of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of Ratings")
plt.show()
```

#### Observation
- Most products with higher ratings (around 4.0–4.5) receive a larger number of customer ratings.
- Products with lower ratings (below 3.0) generally have fewer customer ratings.
- A few highly rated products have extremely large review counts, appearing as outliers.
- Despite this trend, the relationship is weak overall, indicating that a higher rating does not necessarily guarantee more customer ratings.

##### Outlieer Detection

```python
# we try to see if there any outlier using boxplot
plt.figure(figsize=(10,4))

sns.boxplot(data=df1[["ratings","no_of_ratings","discount_price","actual_price"]])
plt.title("Boxplot of Numerical Features")
plt.show()
```

##### Observation
- All numerical features contain outliers, particularly no_of_ratings, discount_price, and actual_price.
- These outliers represent products with exceptionally high review counts or premium prices.
- The ratings column contains very few outliers because ratings are naturally limited to a 1–5 scale.
- Since these outliers correspond to genuine products rather than data errors, they were retained in the dataset.
